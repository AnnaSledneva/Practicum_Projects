# Анализ оттока клиентов банка

**Описание проекта**

Заказчик – отдел маркетинга. Необходимо провести анализ клиентов банка и выделить сегменты клиентов, которые склонны уходить из банка. Результат работы представить в виде презентации для менеджера.

**Цель исследования**

Снизить отток клиентов.

**Описание данных**

- userid — идентификатор пользователя
- score — баллы кредитного скоринга
- city — город
- gender — пол
- age — возраст
- equity — приблизительная оценка собственности клиента
- balance — баланс на счёте
- products — количество продуктов, которыми пользуется клиент
- credit_card — есть ли кредитная карта
- last_activity — был ли клиент активен последнее время
- est_salary — заработная плата клиента
- churn — уходит или нет

## Изучение данных из файла

Импортируем библиотеки, которые пригодятся нам для работы.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from scipy import stats as st

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score,  roc_auc_score, confusion_matrix, plot_confusion_matrix
```

Откроем файл с данными `bank_scrooge.csv` и изучим общую информацию.

```python
data = pd.read_csv('/datasets/bank_scrooge.csv')  
```

```python
display(data.shape)
display(data.head())
data.info()
```

**Вывод:**  
Датафрейм состоит из 10000 наблюдений и 12 полей, из которых 11 признаков и 1 целевая переменная `churn` (факт оттока пользователей). При `сhurn` = 1 считаем, что произошел отток. Есть категориальные признаки - это столбцы `city` и `gender`. Есть пропуски в столбце `balance`. Названия столбцов написаны разным регистром.

## Предобработка данных

### Изменение названий столбцов

Приведем названия столбцов к нижнему регистру:

```python
data.columns = data.columns.str.lower()
data
```

Также для удобства заменим `userid` на `user_id` и `est_salary` на `salary`:

```python
data = data.rename(columns={'userid': 'user_id', 'est_salary': 'salary'})
data
```

### Кодирование признаков

Мы обнаружили, что среди признаков есть категориальные (столбцы `city` и `gender`). Преобразуем их в численные переменные.

С помощью унитарного кодирования (One Hot Encoding) преобразуем значение каждой категории в новый столбец и присвоим столбцу значение 1 или 0 (Истина / Ложь). Чтобы при этом не потерять оригинальный столбец, сначала сделаем его копию.

```python
data['gender_new'] = data['gender']
data['city_new'] = data['city']
data
```

```python
data = pd.get_dummies(data, columns=['gender_new', 'city_new'], drop_first=False)
data
```

### Проверка на дубликаты

Проверим датасет на полные дубликаты:

```python
data.duplicated().sum()
```

Обнаружен 1 полный дубликат. Удалим его:

```python
data.drop_duplicates(inplace=True)
data.duplicated().sum()
```

Посмотрим, сколько у нас всего уникальных пользователей:

```python
data.shape
```

```python
data['user_id'].nunique()
```

Получилось 9927 уникальных пользователя из 9999. Дубликатов очень мало (72), поэтому можно просто избавиться от них.

```python
data.drop_duplicates(subset=['user_id'], inplace = True)
data.shape
```

### Работа с пропусками

Мы обнаружили пропуски в столбце `balance`. Посмотрим, какую долю они составляют:

```python
data['balance'].isna().mean()
```

Получили довольно большой процент пропусков (почти 23 %), просто удалить их нельзя.

```python
data.query('balance.isna()', engine='python')
```

Разделим все данные на 2 группы: первая группа - та, в которой баланс известен (data_balance) и вторая группа - та, в которой баланс не известен (data_no_balance).

```python
data_balance = data.query('balance == balance')
data_balance
```

```python
data_no_balance = data.query('balance != balance')
data_no_balance
```

У нас получилось 2260 клиентов, у которых неизвестен баланс.

Посмотрим на средние значения и медианы признаков при группировке по количеству продуктов (`products`), которыми пользуются пользователи, для этих 2 групп:

```python
display(data_no_balance.groupby('products').agg(['mean', 'median']).T)
data_balance.groupby('products').agg(['mean', 'median']).T
```

```python
no_balance_products = data_no_balance.groupby('products')['user_id'].count().reset_index()
no_balance_products = no_balance_products.rename(columns={'user_id': 'user_count'})
no_balance_products
```

```python
sns.barplot(data=no_balance_products, x='products', y='user_count')
plt.title('График распределения числа продуктов среди пользователей, у которых отсутствует баланс')
plt.xlabel('Количество продуктов')
plt.ylabel('Количество пользователей')
plt.show()
```

```python
balance_products = data_balance.groupby('products')['user_id'].count().reset_index()
balance_products = balance_products.rename(columns={'user_id': 'user_count'})
balance_products
```

```python
sns.barplot(data=balance_products, x='products', y='user_count')
plt.title('График распределения числа продуктов среди пользователей, у которых есть баланс')
plt.xlabel('Количество продуктов')
plt.ylabel('Количество пользователей')
plt.show()
```

```python
sns.histplot(data=data, x='products', hue=data.balance.isna(), multiple='dodge', discrete=True)
plt.show()
```

В группе, в которой не указан баланс, есть 1 пользователь, который вообще не пользуется продуктами банка. А также большинство пользователей (1921) используют только 1 продуктом. В группе, в которой указан баланс, большинство пользователей (4771) имеют 2 продукта, а некоторые даже 4 и 5.

Теперь посмотрим, как различается приблизительная оценка собственности клиента (`equity`) для 2 групп:

```python
no_balance_equity = data_no_balance.groupby('equity')['user_id'].count().reset_index()
no_balance_equity = no_balance_equity.rename(columns={'user_id': 'user_count'})
no_balance_equity
```

```python
sns.barplot(data=no_balance_equity, x='equity', y='user_count')
plt.title('График распределения оценки собственности среди пользователей, у которых отсутствует баланс')
plt.ylabel('Количество пользователей')
plt.xlabel('Оценка собственности')
plt.show()
```

```python
balance_equity = data_balance.groupby('equity')['user_id'].count().reset_index()
balance_equity = balance_equity.rename(columns={'user_id': 'user_count'})
balance_equity
```

```python
sns.barplot(data=balance_equity, x='equity', y='user_count')
plt.title('График распределения оценки собственности среди пользователей, у которых есть баланс')
plt.ylabel('Количество пользователей')
plt.xlabel('Оценка собственности')
plt.show()
```

```python
sns.histplot(data=data, x='equity', hue=data.balance.isna(), multiple="dodge",discrete=True);
```

В группе, в которой не указан баланс, наибольшее число пользователей (2145) вообще не имеет никакой собственности. В другой же группе оценка собственности растет от 0 до 5 вместе с ростом числа пользователей от 409 до 1911. Также есть пользователи с оценками 6, 7, 8, 9, но их всего 271 человек.

Думаю, что пропуски в столбце `balance` нельзя просто заменить на какие-либо значения, они зависят от данных. Мы также не можем их просто отбросить, потому что они составляют почти 23%.

### Изучение средних значений и медиан

1) Посмотрим на средние значения и медианы по всем столбцам:

```python
data.describe()
```

Медиана и среднее сильнее всего отличаются у признаков balance и salary.

2) Посмотрим на средние значения и медианы признаков в двух группах — тех, кто ушел в отток и тех, кто остался:

```python
data.groupby('churn').agg(['mean', 'median']).T
```

```python
new_data = data.drop(columns=['user_id', 'city', 'gender', 'credit_card', 'last_activity', 'gender_new_Ж', 'gender_new_М', 'city_new_Ростов', 'city_new_Рыбинск', 'city_new_Ярославль'])
new_data
```

Посмотрим, насколько средние значения и медианы отточных и неотточных пользователей отличаются от среднего значения и медианы по всем данным.

```python
new_data = new_data.groupby('churn').agg(['mean', 'median'])
new_data
```

Среди всех признаков есть смысл сравнивать значения среднего и медианы только для `score`, `age`, `equity`, `balance`, `products`,  `salary`. Среднее по age, products. Медиана по score, equity и salary.

```python
mean_median = data.drop(columns=['user_id', 'city', 'gender', 'credit_card', 'last_activity', 'gender_new_Ж', 'gender_new_М', 'city_new_Ростов', 'city_new_Рыбинск', 'city_new_Ярославль'])
mean_median = mean_median.agg(['mean', 'median'])
mean_median
```

Посмотрим, по каким показателям отточные и неотточные сильнее всего отличаются.

```python
score = abs(new_data['score'] - mean_median['score'])
score
```

```python
age = abs(new_data['age'] - mean_median['age'])
age
```

```python
equity = abs(new_data['equity'] - mean_median['equity'])
equity
```

```python
balance = abs(new_data['balance'] - mean_median['balance'])
balance
```

```python
products = abs(new_data['products'] - mean_median['products'])
products
```

```python
salary = abs(new_data['salary'] - mean_median['salary'])
salary
```

Почти по всем показателям (`score`, `age`, `equity`, `products`, `salary`) средние значения для отточных и неотточных пользователей различаются почти в 5 раз. Для `balance` средние значения отличаются в 3 раза. Медиана для `score` отличается примерно в 3 раза, для `balance` в 5 раз (это можно объяснить тем, что у нас много пропущенных значений в столбце, которые мы решили не заполнять, поэтому в дальнейшем не будем его рассматривать), по `salary` примерно в 4 раза.

На основании это дальше у нас остается 5 признаков, по которым отточные и неотточные сильнее всего различаются. Это `score`, `age`, `equity`, `products` и `salary`.

### Построение гистограмм

Гистограмма по признаку `balance` получилась неинформативной. Построим новую.

```python
sns.histplot(data=data, x='balance', hue='churn', multiple='dodge', stat='density', common_norm=False)
plt.xlim(0, 6000000)
plt.ylim(0, 0.0000015)
plt.show()
```

Распределение по оттоку сдвинуто вправо, баланс превышает среднее по банку, начиная примерно от 800 тысяч.

```python
sns.histplot(data=data, x='salary', hue='churn', multiple='dodge', stat='density', common_norm=False)
plt.xlim(0, 400000)
plt.show()
```

**Вывод:**  
1) Уровень оттока превышает среднее по банку для значений `score` примерно от 820 до 900.  
2) По возрасту уровень оттока превышает среднее для пользователей 26-30 лет и 50-60.

## Проверка статистических гипотез

### Гипотеза различия дохода между теми клиентами, которые ушли, и теми, которые остались

**Гипотеза №1:** Гипотеза различия дохода (`salary`) между теми клиентами, которые ушли, и теми, которые остались.

Нулевая гипотеза Н_0 - cредний доход клиентов, которые ушли, равен среднему доходу клиентов, которые остались.  
Н_1 - cредний доход клиентов, которые ушли, не равен среднему доходу клиентов, которые остались.

Проведем ttest (st.ttest_ind) и исходя из p-value решим, принять или отвергнуть нулевую гипотезу.

Проанализируем весь столбец `salary`.

```python
data['salary'].describe()
```

Среднее значение отличается от медианы.

```python
sns.histplot(data=data, x='salary')
plt.show()
```

На графике наблюдаются выбросы при значениях `salary` больше 300000. Посмотрим, сколько их.

```python
data['user_id'].count()
```

```python
data.query('salary > 300000')['user_id'].count()
```

Получили 562 значений из 9927. Это примерно 5%. Посмотрим на график при условии salary < 300000:

```python
salary_data = data.query('salary < 300000')
sns.histplot(data=salary_data, x='salary')
plt.show()
```

```python
salary_data['salary'].describe()
```

Распределение уже больше похоже на нормальное, среднее (121742) стало гораздо меньше отличаться от медианы (114713).

Разделим весь столбец дохода на 2 группы: те, кто ушел в отток, и те, кто остался:

```python
group_1 = salary_data.query('churn == 1')
group_1
```

```python
group_2 = salary_data.query('churn == 0')
group_2
```

Посмотрим на распределения двух групп, средние значения и дисперсии. Построим гистограммы:

```python
sns.histplot(data=group_1, x=group_1['salary'], bins=30)
plt.title('Гистограмма распределения дохода пользователей, ушедших в отток')
plt.xlabel('Доход пользователей')
plt.show()
print('Среднее значение для group_1:', group_1['salary'].mean())
print('Дисперсия для платформы group_1:', np.var(group_1['salary'], ddof=1))

sns.histplot(data=group_2, x=group_2['salary'], bins=30)
plt.title('Гистограмма распределения дохода пользователей, не ушедших в отток')
plt.xlabel('Доход пользователей')
plt.show()
print('Среднее значение для group_2:', group_2['salary'].mean())
print('Дисперсия для group_2:', np.var(group_2['salary'], ddof=1))
```

Средние значения почти совпадают, а дисперсии различаются. Добавим в функцию st.ttest_ind аргумент equal_var со значением False.

```python
sample_1 = group_1['salary']
sample_2 = group_2['salary']
 
alpha = .05 # критический уровень статистической значимости
 
results = st.ttest_ind(
    sample_1, 
    sample_2,
    equal_var=False)
 
print('p-значение:', results.pvalue)
 
if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

Я выбрала уровень статистической значимости (alpha) равным 5%, т.к. у нас небольшой размер данных (10000). Если бы их было несколько млн, то можно было взять бы и 1%.

**Вывод:**  
p-value получилось меньше критического уровеня статистической значимости, значит, отвергаем нашу нулевую гипотезу. Получается, что cредние доходы двух групп (ушедших в отток и оставшихся) не совпадают.

Сначала разделим наши данные на обучающую и валидационную выборку функцией train_test_split():

```python
X = data_ml.drop('churn', axis=1)
y = data_ml['churn']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=0
)

X_train.shape, X_test.shape
```

Отдельно определим функцию, которая будет выводить наши метрики:

```python
def print_all_metrics(y_true, y_pred, y_proba, title='Метрики классификации'):
    print(title)
    print('\tAccuracy: {:.2f}'.format(accuracy_score(y_true, y_pred)))
    print('\tPrecision: {:.2f}'.format(precision_score(y_true, y_pred)))
    print('\tRecall: {:.2f}'.format(recall_score(y_true, y_pred)))
    print('\tF1: {:.2f}'.format(f1_score(y_true, y_pred)))
    print('\tROC_AUC: {:.2f}'.format(roc_auc_score(y_true, y_proba)))
```

Теперь попробуем разделить их на группы, объединенные общими признаками.

```python
path = "https://drive.google.com/uc?export=download&id=1-U61mhTz_N1ARjy2XSAZ7IlQqGjeqP0F"
df_temp = pd.read_csv(path)
df_temp.columns = df_temp.columns.str.lower()
examples = [
     ('молодежь', 'age >= 26 and age <= 30'),
     ('люди в возрасте', 'age >= 50 and age <= 60'),
     ('с запасом', 'balance > 800000'),
     ('люди с низким баллом кредитного скоринга', 'score > 820 and score < 900'),
     ('люди со средней оценкой собственности', 'equity >= 3 and equity <= 5'),
     ('обладатели 3-4х продуктов', 'products == 3 or products == 4'),
 ]

#по заданным запросам делаем табличку с размерами разных групп и уровнем оттока в группе и внё её

data_new = []
for text, query in examples:
    this_data = df_temp.query(query)
    other_data_idx = df_temp.index.difference(this_data.index)
    result = this_data\
               .agg({"churn":["count","mean"]}).T
    result.loc['churn','other_churn'] = df_temp.loc[other_data_idx].agg({"churn":"mean"}).item()
    result.index = [text]
    result = result.rename(columns={"count":"num","mean":"churn"})
    data_new.append(result)
    
#придумаем коэффициент, по которому будем ранжировать группы, поправку на размер сегмента можно выбрать и другую:

df_temp = pd.concat(data_new)
df_temp['coef'] = (np.cbrt(df_temp.num) * (df_temp.churn - df_temp.other_churn)).round(1) * 10
df_temp.sort_values('coef',ascending=False)\
        .style.format({"churn":"{:.1%}","other_churn":"{:.1%}","num":"{:.0f}","coef":"{:.0f}"})
```

Выделили группы по признаков, посчитали в них отток. Почему-то на условие 'люди со средней оценкой собственности', 'equity >= 3 and equity =< 5' выдает ошибку, закомментировала его.

## Рекомендации

Чтобы удержать клиентов, можно снизить количество продуктов, которыми они пользуются или объединить несколько продуктов в 1. Для молодежи и людей 50-60 лет нужно повысить баллы кредитного скоринга. Пользователи с балансом более 800 тысяч более склонны к оттоку. Возможно, следует установить какой-то предел на сумму, лежащую на счете, чтобы она не превышала 800 тысяч.
